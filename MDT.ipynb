```python
import pandas as pd
import numpy as np
from tsgm.models.augmentations import MagnitudeWarping

from google.colab import drive
drive.mount('/content/drive')
# Load your dataset from a CSV file
file_path = '/content/drive/MyDrive/Nonfallevents_data.csv'
df = pd.read_csv(file_path)

# Extract the relevant columns
columns = ['AccX', 'AccY', 'AccZ', 'RotX', 'RotY', 'RotZ']
data = df[columns].values

# Reshape data to 3D array (n_samples, n_timesteps, n_features)
# Assuming each sample has 'n_timesteps' time steps
n_timesteps = 100  # Adjust based on your data
n_samples = data.shape[0] // n_timesteps
data = data[:n_samples * n_timesteps].reshape((n_samples, n_timesteps, len(columns)))

# Initialize the Magnitude Warping augmentation model
aug_model = MagnitudeWarping()

# Generate augmented samples
# n_samples specifies the number of augmented samples to generate
# sigma controls the standard deviation of the warping effect
n_augmented_samples = 100  # Adjust based on your requirements
sigma = 0.2
augmented_data = aug_model.generate(X=data, n_samples=n_augmented_samples, sigma=sigma)

# Reshape augmented data back to 2D array for DataFrame
augmented_data = augmented_data.reshape(-1, len(columns))

# Convert the augmented data back to a DataFrame
augmented_df = pd.DataFrame(augmented_data, columns=columns)

# Optionally, concatenate the original and augmented data
expanded_df = pd.concat([df, augmented_df], ignore_index=True)

# Save the expanded dataset to a new CSV file
expanded_file_path = '/content/drive/MyDrive/Nonfall_MDT1_dataset.csv'
expanded_df.to_csv(expanded_file_path, index=False)

print(f"Augmentation complete. Expanded dataset saved to {expanded_file_path}")

```
