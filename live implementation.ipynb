```python
import firebase_admin
from firebase_admin import credentials, db
import pandas as pd
import numpy as np
import pickle
import time
from threading import Timer


from google.colab import drive
drive.mount('/content/drive')


# Initialize Firebase
if not firebase_admin._apps:
    cred = credentials.Certificate('/content/drive/MyDrive/CaneConnect/JSON file.json')
    firebase_admin.initialize_app(cred, {'databaseURL': 'https://caneconnect-a0ed2-default-rtdb.asia-southeast1.firebasedatabase.app/'})
    print("Firebase initialized:", firebase_admin._apps)
else:
  print("Firebase app is already initialized.")


print("Connected to Firebase. Model and scaler loaded successfully!")


# Load the trained MLP model (saved as .pkl)
with open("model.pkl", "rb") as f:
    model = pickle.load(f)


# Load the saved scaler (from training)
with open("scaler.pkl", "rb") as f:
    scaler = pickle.load(f)


# Function to fetch live sensor data from Firebase
def fetch_live_sensor_data():
    ref = db.reference("Sensor")  # Fetch data from 'Sensor' node
    sensor_data = ref.get()


    if sensor_data:
        df_live = pd.DataFrame([sensor_data])  # Convert dict to DataFrame
        print("\n Live Sensor Data Fetched:")
        print(df_live)


        # Select only required features for prediction
        features = ["accX", "accY", "accZ", "rotX", "rotY", "rotZ"]
        X_live = df_live[features].values


        # Apply the same scaling used in training
        X_live_scaled = scaler.transform(X_live)
        return X_live_scaled


    print(" No sensor data available in Firebase.")
    return None


# Function to predict falls using the MLP model
def predict_fall(X_live_scaled):
    if X_live_scaled is not None:
        prediction = model.predict(X_live_scaled)  # Predict fall (1) or non-fall (0)
        print(f"\n Model Prediction: {prediction[0]}")


        return prediction
    return None


alert_sent=False
while True:
    # Step 1: Fetch live sensor data & preprocess
    X_live_scaled = fetch_live_sensor_data()


    # Step 2: Predict Fall or No Fall
    fall_prediction = predict_fall(X_live_scaled)


    if fall_prediction is not None:
        if fall_prediction[0] == 1:
            print("\nFALL DETECTED! (High Risk)")
            if not alert_sent:
                sendAlert()
                alert_sent = True
            else:
                print("Alert already sent. Waiting for cooldown period to end.")

        else:
            print("\nNo Fall (Safe Condition)")
            alert_sent=False
```
